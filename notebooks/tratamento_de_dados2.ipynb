## Dia 6

* Executar pipeline de integração de dados de empréstimos e alunos
    * Ler, tratar e limpar os dados em xls
    * Ler , tratar e limpar os dados em json
    * unir os dados da base em xls e em json
    * Ler dados de empréstimos tratados
    * salvar os dados num novo arquivo



```python
# imports

import pandas as pd
import json
```

```python
# lendo os dados de matrículas dos alunos
# necessário ter biblioteca openpyxl instalada: pip install openpyxl

data_xls_before_2010 = pd.read_excel('../datasets/dados_alunos/matricula_alunos.xlsx',engine='openpyxl', sheet_name='Até 2010')
data_xls_before_2010
```

```python
data_xls_after_2010 = pd.read_excel('../datasets/dados_alunos/matricula_alunos.xlsx',engine='openpyxl', sheet_name='Após 2010')
data_xls_after_2010
```

```python
# limpando o cabeçalho duplicado da tabela data_xls_before_2010

data_xls_before_2010 = data_xls_before_2010[1:]
data_xls_before_2010.columns = ['matricula', 'vinculo','curso']
```

```python
data_xls_before_2010
```

```python
# limpando o cabeçalho duplicado da tabela data_xls_after_2010

data_xls_after_2010 = data_xls_after_2010[1:]
data_xls_after_2010.columns = ['matricula', 'vinculo','curso']  
```

```python
data_xls_after_2010
```

```python
# unindo os dataframes xls em um único dataframe

data_xls = pd.concat([data_xls_before_2010, data_xls_after_2010], ignore_index=True)
data_xls
```

```python
data_xls.info()
```

```python
# eliminando o ponto flutuante da coluna matricula

data_xls.loc[:,'matricula'] = data_xls['matricula'].str.replace('.0','')  
data_xls
```

```python
# Verificando se existem dados faltantes

data_xls.isnull().sum()
```

```python
# verificando se existem dados duplicados

data_xls.duplicated().sum()
```

```python
# Lendo os dados do arquivo JSON


data_json = pd.read_json('../datasets/dados_alunos/matricula_alunos2.json')
data_json.head()
```

```python
# Parseando a variável string JSON  para DataFrame

stundents_grad = data_json['registros'][0]
stundents_post_grad = data_json['registros'][1]
data_json_grad = pd.json_normalize(json.loads(stundents_grad)) # 
data_json_post_grad = pd.json_normalize(json.loads(stundents_post_grad)) #
df_json = pd.concat([data_json_grad, data_json_post_grad], ignore_index=True)
df_json.head()
```

```python
# analisando os dados do DataFrame

df_json.info()
```

```python
# eliminando o ponto flutuante da coluna matricula

df_json.loc[:,'matricula_ou_siape'] = df_json['matricula_ou_siape'].str.replace('.0','')
df_json
```

```python
# unindo os dois DataFrames
data_xls.columns = ['matricula_ou_siape', 'vinculo', 'curso']
df_json.columns = data_xls.columns

df_all = pd.concat([data_xls, df_json], ignore_index=True)
df_all = df_all[['matricula_ou_siape', 'curso']]
df_all
```

```python
# verificando se tem dados duplicados no DataFrame final
df_all.duplicated().sum()
```

```python
# verificando dados faltantes no DataFrame unido
df_all.isnull().sum()
```

```python
# verificando os cursos da base de dados unida
df_all['curso'].value_counts()
```

```python
# lendo os dados de empréstimos de exemplares

df_data_loans = pd.read_parquet('../datasets/dados_tratados/dados_emprestimos_tratados.parquet')
df_data_loans.head()
```

```python
# unindo os dados de alunos com os dados de empréstimos

df_merged = pd.merge(df_data_loans, df_all, how='left', left_on='matricula_ou_siape', right_on='matricula_ou_siape')
df_merged.head()
```

```python
# vericando os dados faltantes no DataFrame final
df_merged.isnull().sum()
```

```python
# verificando dados duplicados no DataFrame final
df_merged.duplicated().sum()
```

```python
# salvando os dados tratados em um arquivo parquet

df_merged.to_parquet('../datasets/dados_tratados/dados_emprestimos_com_alunos.parquet', index=False)
```
