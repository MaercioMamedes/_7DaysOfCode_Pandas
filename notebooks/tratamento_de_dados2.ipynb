```python
# Lendo os dados do arquivo JSON


data_json = pd.read_json('../datasets/dados_alunos/matricula_alunos2.json')
data_json.head()
```

```python
# Parseando a variável string JSON  para DataFrame

stundents_grad = data_json['registros'][0]
df_data_json_grad = pd.json_normalize(json.loads(stundents_grad)) # 
df_data_json_grad
```

```python
# analisando os dados do DataFrame

df_data_json_grad.info()
```

```python
# eliminando o ponto flutuante da coluna matricula

df_data_json_grad.loc[:,'matricula_ou_siape'] = df_data_json_grad['matricula_ou_siape'].str.replace('.0','')
df_data_json_grad
```

```python
# unindo os dois DataFrames
data_xls_grad.columns = ['matricula_ou_siape', 'vinculo', 'curso']
df_data_json_grad.columns = data_xls_grad.columns

df_all = pd.concat([data_xls_grad, df_data_json_grad], ignore_index=True)
df_all
```

```python
# verificando se tem dados duplicados no DataFrame final
df_all.duplicated().sum()
```

```python
# verificando dados faltantes no DataFrame unido
df_all.isnull().sum()
```

```python
# verificando os cursos da base de dados unida
df_all['curso'].value_counts()
```

```python
# lendo os dados de empréstimos de exemplares

df_data_loans = pd.read_parquet('../datasets/dados_tratados/dados_emprestimos_tratados.parquet')
df_data_loans.head()
```

```python
# unindo os dados de alunos com os dados de empréstimos

df_merged = pd.merge(df_data_loans, df_all, how='left', left_on='matricula_ou_siape', right_on='matricula_ou_siape')
df_merged.head()
```

```python
# selecionando apenas os alunos de graduação

df_merged_grad = df_merged[df_merged['vinculo']=='ALUNO DE GRADUAÇÃO']
df_merged_grad.head()
```

```python
# excluindo colunas desnecessárias
df_merged_grad = df_merged_grad.drop(columns=['vinculo'])
df_merged_grad.head()
```

```python
# vericando os dados faltantes no DataFrame final
df_merged_grad.isnull().sum()
```

```python
# verificando dados duplicados no DataFrame final
df_merged_grad.duplicated().sum()
```

```python
# transformando a coluna da data para ano de empréstimo
df_merged_grad['data_emprestimo'] = pd.to_datetime(df_merged_grad['data_emprestimo']).dt.year
```

```python
# selecionando os empréstimos apenas dos cursos avaliados no período 2015 e 2020

df_merged_grad = df_merged_grad.query("""curso == ['BIBLIOTECONOMIA', 'CIÊNCIAS SOCIAIS', 'COMUNICAÇÃO SOCIAL', 'DIREITO', 'FILOSOFIA', 'PEDAGOGIA'] & (data_emprestimo >= 2015) """)

df_merged_grad.head()
```

```python
# selecionando as colunas relevantes para a análise final
df_final = df_merged_grad[['id_emprestimo', 'curso', 'data_emprestimo']]
df_final.head()
```

```python
df_final.iloc[:,[1,2]].value_counts()
```

```python
# agrupando a quantidade de empréstimos por curso e ano
df_grouped = df_final.groupby(['curso', 'data_emprestimo'])['id_emprestimo'].count().reset_index()
df_grouped.rename(columns={'id_emprestimo':'quantidade_emprestimos'}, inplace=True)
df_grouped
```

```python
# transformando ano em colunas
df_pivot = df_grouped.pivot_table(
    index='curso', 
    columns='data_emprestimo', 
    values='quantidade_emprestimos',
    #fill_value='-',
    aggfunc='sum',
    margins=True,
    margins_name='Total')
df_pivot
```
