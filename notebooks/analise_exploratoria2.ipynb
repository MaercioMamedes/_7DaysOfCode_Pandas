- calcular a quantidade de empréstimos realizados entre 2015 e 2020 por cada curso de graduação que passará pela avaliação.
- Cursos a serem avaliados:
    * Biblioteconomia
    * Ciências sociais
    * Comunicação social
    * Direito
    * Filosofia
    * Pedagogia
- Formato da tabela

Gerar uma tabela com as seguintes características:

Índice: Cursos
Colunas: Ano
Valores: Quantidade de empréstimos
Total: Acrescente uma linha e uma coluna de total a tabela

| Índice | Ano | Total |
|----------|----------|----------|
| Cursos | qtd de empréstimo  | Total de emprestimos  |

```python
# imports

import pandas as pd
```

```python
# lendo os arquivos tratados com os dados do alunos

data = pd.read_parquet('../datasets/dados_tratados/dados_emprestimos_com_alunos.parquet')
data.head()
```

```python
# selecionando apenas os alunos de graduação

df_grad = data[data['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO']
df_grad.head()
```

```python
# selecionando as colunas relevantes para a análise final
df_grad = df_grad[['id_emprestimo', 'curso', 'data_emprestimo']]
df_grad.head()
```

```python
# transformando a coluna data_emprestimo para valores em ano

df_grad['data_emprestimo'] = pd.to_datetime(df_grad['data_emprestimo']).dt.year
df_grad
```

```python
# selecionando os cursos e o período a serem avaliados

df_grad = df_grad.query("""curso == ['BIBLIOTECONOMIA', 'CIÊNCIAS SOCIAIS', 'COMUNICAÇÃO SOCIAL', 'DIREITO', 'FILOSOFIA', 'PEDAGOGIA'] & (data_emprestimo >= 2015) """)

df_grad.head()
```

```python
# agrupando a quantidade de empréstimos por curso e ano

df_grad = df_grad.groupby(['curso', 'data_emprestimo'])['id_emprestimo'].count().reset_index()
df_grad.rename(columns={'id_emprestimo':'quantidade_emprestimos'}, inplace=True)
df_grad
```

```python
# pivotando a tabela para melhor visualização dos dados

df_pivot = df_grad.pivot_table(
    index='curso', 
    columns='data_emprestimo', 
    values='quantidade_emprestimos',
    #fill_value='-',
    aggfunc='sum',
    margins=True,
    margins_name='Total')
df_pivot
```
